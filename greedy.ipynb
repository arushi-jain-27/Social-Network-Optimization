```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("soc-sign-bitcoinotc.csv", header=None)
n = max(max(df[0]), max(df[1]))
A = np.zeros([n,n], dtype=int)
for index, row in df.iterrows():
    if int(row[2]) > 0:
        A[int(row[0])-1, int(row[1])-1] = int(row[2])
A  
```

```python
synthetic = pd.read_csv("synthetic.csv")
budget = synthetic["Budget"]
synthetic["Value"] = synthetic["Total Influencing"]/synthetic["Budget"]
```

```python
threshold = synthetic["Threshold"]
```

```python
synthetic
```

```python
selected = set()
budget = 1000000
seedCount = 5
i = 0
synthetic = synthetic.sort_values(by="Value", ascending=False, ignore_index=True)
while budget and i < n:
    if budget > synthetic.at[i, "Budget"]:
        selected.add(synthetic.at[i, "Node"])
        print(i, synthetic.at[i, "Node"])
        budget -= synthetic.at[i, "Budget"]
    i += 1
selected    
```

```python
synthetic[synthetic["Node"].isin(list(selected))]
```

```python
T = 2
count = 0
x = np.zeros([n,n], dtype=int)
for node in selected:
    x[:,node] = A[:, node]
    count += 1
for t in range(2,T):
    for i in range(n):
        if sum(x[i,:]) >= threshold[i]:
            x[:, i] = A[:, i]
            count += 1
```
